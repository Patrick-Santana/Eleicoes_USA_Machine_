# Stopwords

- S√£o palavras ou termos de uso frequente que n√£o possuem relev√¢ncia nas seten√ßas.

- Exemplos: A, de, do, para , etc .

```python
def RemoveStopWords(instancia):
  stopwords = set(nltk.corpus.stopwords('portuguese'))
  palavras = [i for i in instancia.split() if not i in stopwords]

  return (" ".join(palavras))
```

# Stemming

- t√©cnica Utilizada para converter palavras em seu radical.
- Ex: ilegal -> ileg.

```python
def Stemming(instancia):
  stemmer = nltk.stem.RSLPStemmer()
  palavras = []

  for w in instancia.split():
    palavras.append(stemmer.stem(w))
  return (" ".join(palavras))
```
